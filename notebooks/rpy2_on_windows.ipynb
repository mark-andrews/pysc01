```python
from rpy2.robjects import pandas2ri
from rpy2.robjects.packages import importr
import rpy2.robjects as ro
R = ro.r
pandas2ri.activate()
base = importr('base')
```

```python
import pandas as pd
```

```python
weight_df = pd.read_csv("https://raw.githubusercontent.com/mark-andrews/pysc01/master/data/weight.csv")
```

```python
M = R.lm('weight ~ height', data = weight_df)
```

```python
R.coef(M)
```

```python
print(base.summary(M))
```

```python
print(base.summary(M).rx2('coefficients'))
```
