```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
weight_df = pd.read_csv("https://raw.githubusercontent.com/mark-andrews/pysc01/master/data/weight.csv")
```

```python
weight_df
```

```python
sns.scatterplot(x = 'height', y = 'weight', data = weight_df)
```

```python
sns.scatterplot(x = 'height', y = 'weight', hue = 'gender', data = weight_df)
```

```python
sns.relplot(x = 'height',
            y = 'weight',
            hue = 'gender',
            col = 'race',
            col_wrap = 3,
            data = weight_df)
```

```python
sns.lmplot(x = 'height',
           y = 'weight',
           hue = 'gender',
           col = 'race',
           col_wrap = 3,
           data = weight_df)
```

```python
%rep 82
```

```python
%rep 81
```

```python
sns.lmplot(x = 'height',
           y = 'weight',
           hue = 'gender',
           col = 'race',
           scatter_kws = {'s': 5, 'alpha': 0.5},
           col_wrap = 3,
           ci = None,
           data = weight_df)
```

```python
sns.catplot(x = 'gender', y = 'height', ci = 'sd', kind = 'bar', data = weight_df)
```

```python
from plotnine import ggplot, aes, geom_point, stat_smooth, facet_wrap, theme_classic
```
