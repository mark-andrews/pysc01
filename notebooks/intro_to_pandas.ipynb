## pivots

```python
df_1 = pd.DataFrame(
    {
        'name': ['John', 'Jane', 'Mary', 'John', 'Jane', 'Mary'],
        'groupvar': ['A', 'A', 'A', 'B', 'B', 'B'],
        'measure': [2, 11, 1, 3, 4, 5]
    }
)
```

```python
df_1
```

```python
df_2 = df_1.pivot(index = 'name', columns = 'groupvar', values = 'measure').reset_index()
```

```python
df_2.columns
```

```python
df_2
```

```python
df_2.melt(id_vars = 'name')
```
